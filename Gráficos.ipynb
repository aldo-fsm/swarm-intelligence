```python
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
data_dir = '/home/aldo/Área de Trabalho/'
results_sphere = np.load(data_dir+'results_sphere.npy')
results_rastrigins = np.load(data_dir+'results_rastrigins.npy')
results_rosenbrock = np.load(data_dir+'results_rosenbrock.npy')
```

```python
labels = ['Básico', 
          'Inercia - constante\nTopologia Global', 'Inercia - constante\nTopologia Local', 'Inercia - constante\nTopologia Focal',
          'Inercia - decaimento linear\nTopologia Global', 'Inercia - decaimento linear\nTopologia Local', 'Inercia - decaimento linear\nTopologia Focal',
          'Constrição de Clerc\nTopologia Global', 'Constrição de Clerc\nTopologia Local', 'Constrição de Clerc\nTopologia Focal'
         ]
```

```python
list(enumerate(labels))
```

```python
figure_dir = 'figures/Prática N1/'
```

```python
plt.rcParams['figure.figsize'] = 10, 7
```

```python
plt.grid()
plt.title("Boxplot - Sphere Function")
plt.boxplot(results_sphere[:, :, -1]);
plt.xticks(range(1,1+len(results_sphere[0, :])), labels, rotation='vertical')
plt.xlabel('Algoritmo')
plt.ylabel('Fitness')
plt.tight_layout()
plt.savefig(os.path.join(figure_dir, 'Boxplot - Sphere.png'))
```

```python
plt.grid()
plt.title("Boxplot - Rastrigin's Function")
plt.boxplot(results_rastrigins[:, :, -1]);
plt.xticks(range(1,1+len(results_sphere[0, :])), labels, rotation='vertical')
plt.xlabel('Algoritmo')
plt.ylabel('Fitness')
plt.tight_layout()
plt.savefig(os.path.join(figure_dir, 'Boxplot - Rastrigins.png'))
```

```python
plt.grid()
plt.title("Boxplot - Rosenbrock Function")
plt.boxplot(results_rosenbrock[:, :, -1]);
plt.xticks(range(1,1+len(results_sphere[0, :])), labels, rotation='vertical')
plt.xlabel('Algoritmo')
plt.ylabel('Fitness')
plt.tight_layout()
plt.savefig(os.path.join(figure_dir, 'Boxplot - Rosenbrock.png'))
```

```python
def filled_curve(data, color=None, marker=None, markevery=None): # input shape (num_experiments, num_iterations)
    mean_line = plt.plot(np.mean(data, axis=0), color=color, marker=marker, markevery=markevery)[0]
    plt.fill_between(range(data.shape[1]), np.min(data, axis=0), np.max(data, axis=0), alpha=0.1, color=mean_line.get_color())
```

```python
plt.title('Convergência - Sphere Function')
plt.grid()
for i, label in enumerate(labels):
    filled_curve(results_sphere[:,i,:])
plt.legend(labels, loc='best', ncol=3)
plt.xlabel('Iterações')
plt.ylabel('Fitness')
# plt.xlim([-2, 600])
plt.savefig(os.path.join(figure_dir, 'Convergence - Sphere.png'))
```

```python
plt.title("Convergência - Rastrigin's Function")
plt.grid()
for i, label in enumerate(labels):
    filled_curve(results_rastrigins[:,i,:])
plt.legend(labels, loc='best', ncol=3)
plt.xlabel('Iterações')
plt.ylabel('Fitness')
# plt.xlim([-2, 600])
plt.savefig(os.path.join(figure_dir, 'Convergence - Rastrigins.png'))
```

```python
plt.title('Convergência - Rosenbrock Function')
plt.grid()
for i, label in enumerate(labels):
    filled_curve(results_rosenbrock[:,i,:])
plt.legend(labels, loc='best', ncol=3)
plt.xlabel('Iterações')
plt.ylabel('Fitness')
# plt.xlim([-2, 600])
plt.savefig(os.path.join(figure_dir, 'Convergence - Rosenbrock.png'))
```
