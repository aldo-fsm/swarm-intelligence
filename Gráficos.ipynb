```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data_dir = '/home/aldo/Área de Trabalho/'
results_sphere = np.load(data_dir+'results_sphere.npy')
results_rastrigins = np.load(data_dir+'results_rastrigins.npy')
results_rosenbrock = np.load(data_dir+'results_rosenbrock.npy')
```

```python
labels = ['Básico', 
          'Inercia - constante\nGlobal', 'Inercia - constante\nLocal', 'Inercia - constante\nFocal',
          'Inercia - decaimento linear\nGlobal', 'Inercia - decaimento linear\nLocal', 'Inercia - decaimento linear\nFocal',
          'Constrição de Clerc\nGlobal', 'Constrição de Clerc\nLocal', 'Constrição de Clerc\nFocal'
         ]
```

```python
figure_dir = 'figures/Prática N1/boxplots/'
```

```python
plt.grid()
plt.title("Sphere Function")
plt.boxplot(results_sphere[:, :, -1]);
plt.xticks(range(1,1+len(results_sphere[0, :])), labels, rotation='vertical')
plt.xlabel('Algoritmo')
plt.ylabel('Fitness')
plt.savefig(figure_dir+'Boxplot - Sphere.png')
```

```python
plt.grid()
plt.title("Rastrigin's Function")
plt.boxplot(results_rastrigins[:, :, -1]);
plt.xticks(range(1,1+len(results_sphere[0, :])), labels, rotation='vertical')
plt.xlabel('Algoritmo')
plt.ylabel('Fitness')
plt.savefig(figure_dir+'Boxplot - Rastrigins.png')
```
